# Data Exploration (Exploratory Data Analysis – EDA)

Purpose:
	•	To understand the structure and characteristics of the dataset.
	•	Helps to discover patterns, trends, and anomalies before applying statistical or machine learning models.
	•	Guides in choosing the right analysis methods.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
df = pd.read_csv("data/processed/student_performance_cleaned.csv")
```

```python
# Inspect data
df.head()
df.info()
df.describe()
```

```python
# Shape of the data
print(df.shape)  # rows, columns

# Column names
print(df.columns)

# Info & types
print(df.info())

# Peek at the first few rows
print(df.head())

# Peek at the last few rows
print(df.tail())
```

```python
# Summary for numeric columns
print(df.describe())

# Summary for categorical columns
print(df.describe(include='object'))
```
